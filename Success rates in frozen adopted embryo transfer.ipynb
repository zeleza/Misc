## Assumptions of modeling below

 - I am assuming that one will keep trying to achieve a pregnancy until one has a successful live birth
 - I am assuming that the success of any given cycle is independent of the success of another. This means that the distribution of success is geometric
 - **The assumption that the success of one cycle being independent of another is probably false. Therefore, the true probability is probably lower!**
 - I am making no effort to account for various variables that are probably important, such as the age of the recipient, age of donor, freezing conditions of egg, thawing conditions of egg, etc.
 - Actual probability of success if probably lower than reported

I have seen ranges of probability of success per cycle from 10% to 35% for frozen embryo transfer resulting in life birth, so I'm just plotting the success rates for that range

*Unfortunately, as a 47 year old woman with a few unsuccessful attempts, I would be looking at the lower end of the probability distributions*

```r
# importing graphing libraries
options(warn=-1)
library(ggplot2)
library(tidyverse)
```

```r
# 5% success
prob = 0.05 # assuming 15% probability
n_cycles <- c(1:20) # examining 1-15 cycles
cdf_births <- 100*pgeom((1:20), prob = prob) # calculating probability distribution
success_df <- data.frame(n_cycles,cdf_births)
average <- (1-prob)/prob

title = sprintf("Assuming %.0f percent success rate: average cycles needed = %.1f",(prob*100),average)

pc_5 <- ggplot(data=success_df, aes(x=n_cycles,y=cdf_births)) + geom_bar(stat="identity") + xlab("Number of cycles") +
        ylab("Cumulative probability of success (%)") + ggtitle(title) +
        geom_vline(xintercept = average, linetype = "dashed", color = "red", size = 2) +
        scale_x_continuous(breaks = seq(1,20,1))

# 10% success
prob = 0.10 # assuming 15% probability
n_cycles <- c(1:15) # examining 1-15 cycles
cdf_births <- 100*pgeom((1:15), prob = prob) # calculating probability distribution
success_df <- data.frame(n_cycles,cdf_births)
average <- (1-prob)/prob

title = sprintf("Assuming %.0f percent success rate: average cycles needed = %.1f",(prob*100),average)

pc_10 <- ggplot(data=success_df, aes(x=n_cycles,y=cdf_births)) + geom_bar(stat="identity") + xlab("Number of cycles") +
        ylab("Cumulative probability of success (%)") + ggtitle(title) +
        geom_vline(xintercept = average, linetype = "dashed", color = "red", size = 2) +
        scale_x_continuous(breaks = seq(1,15,1))

# 15% success
prob = 0.15 # assuming 15% probability
n_cycles <- c(1:15) # examining 1-15 cycles
cdf_births <- 100*pgeom((1:15), prob = prob) # calculating probability distribution
success_df <- data.frame(n_cycles,cdf_births)
average <- (1-prob)/prob

title = sprintf("Assuming %.0f percent success rate: average cycles needed = %.1f",(prob*100),average)

pc_15 <- ggplot(data=success_df, aes(x=n_cycles,y=cdf_births)) + geom_bar(stat="identity") + xlab("Number of cycles") +
        ylab("Cumulative probability of success (%)") + ggtitle(title) +
        geom_vline(xintercept = average, linetype = "dashed", color = "red", size = 2) +
        scale_x_continuous(breaks = seq(1,15,1))

# 20% success
prob = 0.20 # assuming 15% probability
n_cycles <- c(1:15) # examining 1-15 cycles
cdf_births <- 100*pgeom((1:15), prob = prob) # calculating probability distribution
success_df <- data.frame(n_cycles,cdf_births)
average <- (1-prob)/prob

title = sprintf("Assuming %.0f percent success rate: average cycles needed = %.1f",(prob*100),average)

pc_20 <- ggplot(data=success_df, aes(x=n_cycles,y=cdf_births)) + geom_bar(stat="identity") + xlab("Number of cycles") +
        ylab("Cumulative probability of success (%)") + ggtitle(title) +
        geom_vline(xintercept = average, linetype = "dashed", color = "red", size = 2) +
        scale_x_continuous(breaks = seq(1,15,1))

# 25% success
prob = 0.25 # assuming 15% probability
n_cycles <- c(1:15) # examining 1-15 cycles
cdf_births <- 100*pgeom((1:15), prob = prob) # calculating probability distribution
success_df <- data.frame(n_cycles,cdf_births)
average <- (1-prob)/prob

title = sprintf("Assuming %.0f percent success rate: average cycles needed = %.1f",(prob*100),average)

pc_25 <- ggplot(data=success_df, aes(x=n_cycles,y=cdf_births)) + geom_bar(stat="identity") + xlab("Number of cycles") +
        ylab("Cumulative probability of success (%)") + ggtitle(title) +
        geom_vline(xintercept = average, linetype = "dashed", color = "red", size = 2) +
        scale_x_continuous(breaks = seq(1,15,1))

# 30% success
prob = 0.30 # assuming 15% probability
n_cycles <- c(1:15) # examining 1-15 cycles
cdf_births <- 100*pgeom((1:15), prob = prob) # calculating probability distribution
success_df <- data.frame(n_cycles,cdf_births)
average <- (1-prob)/prob

title = sprintf("Assuming %.0f percent success rate: average cycles needed = %.1f",(prob*100),average)

pc_30 <- ggplot(data=success_df, aes(x=n_cycles,y=cdf_births)) + geom_bar(stat="identity") + xlab("Number of cycles") +
        ylab("Cumulative probability of success (%)") + ggtitle(title) +
        geom_vline(xintercept = average, linetype = "dashed", color = "red", size = 2) +
        scale_x_continuous(breaks = seq(1,15,1))

# 35% success
prob = 0.35 # assuming 15% probability
n_cycles <- c(1:15) # examining 1-15 cycles
cdf_births <- 100*pgeom((1:15), prob = prob) # calculating probability distribution
success_df <- data.frame(n_cycles,cdf_births)
average <- (1-prob)/prob

title = sprintf("Assuming %.0f percent success rate: average cycles needed = %.1f",(prob*100),average)

pc_35 <- ggplot(data=success_df, aes(x=n_cycles,y=cdf_births)) + geom_bar(stat="identity") + xlab("Number of cycles") +
        ylab("Cumulative probability of success (%)") + ggtitle(title) +
        geom_vline(xintercept = average, linetype = "dashed", color = "red", size = 2) +
        scale_x_continuous(breaks = seq(1,15,1))
```
